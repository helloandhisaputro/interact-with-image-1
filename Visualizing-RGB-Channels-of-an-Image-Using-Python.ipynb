```python
import matplotlib.pyplot as plt
from skimage import io

# Nama file gambar lokal
image_path = 'dragon-ball-daima-button-1697147664709.jpg'

# Membaca gambar dari file lokal
img = io.imread(image_path)

# Membuat subplot dengan ukuran 10x7
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 7))

# Mengatur layout subplot agar tidak tumpang tindih
plt.tight_layout(pad=2.0)

# Gambar asli
axes[0, 0].imshow(img)
axes[0, 0].axis('off')
axes[0, 0].set_title("Original Image")

# Kanal warna merah (R)
axes[0, 1].imshow(img[:, :, 0], cmap='gray')
axes[0, 1].axis('off')
axes[0, 1].set_title("Red Channel")

# Kanal warna hijau (G)
axes[1, 0].imshow(img[:, :, 1], cmap='gray')
axes[1, 0].axis('off')
axes[1, 0].set_title("Green Channel")

# Kanal warna biru (B)
axes[1, 1].imshow(img[:, :, 2], cmap='gray')
axes[1, 1].axis('off')
axes[1, 1].set_title("Blue Channel")

# Menampilkan plot
plt.show()
```
